**Selection of texts 1**

```python
from pytesseract import Output
```

```python
img = cv2.imread('/content/test01.jpg')
rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2_imshow(rgb)
```

```python
!wget -O ./tessdata/eng.traineddata https://github.com/tesseract-ocr/tessdata/blob/main/eng.traineddata?raw=true
```

```python
config_tesseract = '--tessdata-dir tessdata'
result = pytesseract.image_to_data(rgb,lang='eng',config=config_tesseract,output_type=Output.DICT)
print(result)
```

**Selection of texts 2**

```python
print(result['text'],len(result['text']))
```

```python
def bounding_box(result,img,i,color=(255,100,6)):
  x = result['left'][i]
  y = result['top'][i]
  w = result['width'][i]
  h = result['height'][i]
  cv2.rectangle(img,(x,y),(x+w,y+h),color,2)
  return x,y,img
```

```python
min_confidence = 40
img_copy = rgb.copy()
for i in range(0,len(result['text'])):
  #print(i)
  confidence = int(result['conf'][i])
  #print(confidence)
  if(confidence>min_confidence):
    #print(confidence)
    x,y,img = bounding_box(result,img_copy,i)
    #print(x,y)
    text = result['text'][i]
    cv2.putText(img_copy,text,(x,y-10),cv2.FONT_HERSHEY_SIMPLEX,1.1,(0,0,255))
cv2_imshow(img_copy)

```

```python
img2 = cv2.imread('/content/test02-02.jpg')
rgb = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
cv2_imshow(rgb)
```

```python
config_tesseract = '--tessdata-dir tessdata'
result = pytesseract.image_to_data(rgb,lang='por',config=config_tesseract,output_type=Output.DICT)
print(result)
```

```python
from PIL import ImageFont,ImageDraw,Image
font = '/content/calibri.ttf'
```

```python
def write_text(text,x,y,img,font,font_size=12):
  font = ImageFont.truetype(font,font_size)
  img_pil = Image.fromarray(img)
  draw = ImageDraw.Draw(img_pil)
  draw.text((x,y-font_size),text,font = font)
  img = np.array(img_pil)
  return img
```

```python
min_confidence = 40
img_copy = rgb.copy()
for i in range(0,len(result['text'])):
  #print(i)
  confidence = int(result['conf'][i])
  #print(confidence)
  if(confidence>min_confidence):
    #print(confidence)
    x,y,img = bounding_box(result,img_copy,i)
    #print(x,y)
    text = result['text'][i]
    img_copy = write_text(text,x,y,img_copy,font)
cv2_imshow(img_copy)
```

**Search using regular expression**

```python
import re
```

```python
img = cv2.imread('/content/table_test.jpg')
rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2_imshow(rgb)
```

```python
config_tesseract = '--tessdata-dir tessdata'
result = pytesseract.image_to_data(rgb,lang='por',config=config_tesseract,output_type=Output.DICT)
print(result)
```

```python
date_pattern = '^(0[1-9]|[12][0-9]|3[01])/(0[1-9]|[012])/(19|20)\d\d$'
```

```python
dates =[]
min_confidence = 40
img_copy = rgb.copy()
for i in range(0,len(result['text'])):
  confidence = int(result['conf'][i])
  if(confidence>min_confidence):
     text = result['text'][i]
     if re.match(date_pattern,text):
        x,y,img = bounding_box(result,img_copy,i,(0,0,255))
        img_copy = write_text(text,x,y,img_copy,font,12)
        dates.append(text)
     else:
        x,y, img_copy = bounding_box(result,img_copy,i)

cv2_imshow(img_copy)
```
